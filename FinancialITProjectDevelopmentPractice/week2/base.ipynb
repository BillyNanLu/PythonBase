```python
# 抢苹果，AB分别抛硬币。A先抛硬币假如抛出正面则A吃苹果，假如抛出反面则给B抛；B假如抛出反面则吃苹果，假如抛出正面则给A抛。则A吃到的苹果概率是多少？
import random
from random import randint

from sympy import false


def once():
    turn = 'A'
    while True:
        toss = random.choice([0, 1])    # 1 = head, 0 = tail
        if turn == 'A':
            if toss == 1:
                return 'A'
                break
            else:
                turn = 'B'
        else:
            if toss == 0:
                return 'B'
                break
            else:
                turn = 'A'

# 模拟很多次
N = 100000
a_wins = 0
for _ in range(N):
    if once() == 'A':
        a_wins += 1

p_a = a_wins / N
p_b = 1 - p_a

print(f"A吃到苹果概率：{p_a:.4f}")
print(f"B吃到苹果概率：{p_b:.4f}")
```

```python
# 计算PI
import random

print(f"在正方形内随机生成一个点P({x}, {y})")

TOTAL = 100000
success = 0
fail = 0

for _ in range(TOTAL):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)

    if (x**2 + y**2)**(1/2) <= 1:
        # print(f"点P({x}, {y})在圆内")
        success += 1
    else:
        # print(f"点P({x}, {y})在圆外")
        fail += 1

print(f"pi = {success/TOTAL * 4}")
```

```python
# 练习1：录入唐诗一首，每句唐诗随机扣出一个字，用_替代，然后出一个唐诗填空的题目。根据回答给出得分。\
import random

# 春思【唐】李白
poem_lines = ['燕草如碧丝', '秦桑低绿枝', '当君怀归日', '是妾断肠时', '春风不相识', '何事入罗帏']

questions = []
answers = []
for line in poem_lines:
    idxs = [i for i, ch in enumerate(line) if ch.strip() != '']
    if not idxs:
        questions.append(line)
        answers.append("")
        continue

    idx = random.choice(idxs)
    blank_char = line[idx]
    question_line = line[:idx] + '_' + line[idx+1:]
    questions.append(question_line)
    answers.append(blank_char)
print(questions)
print(answers)

print("\n=== 填空题 ===")
for i, q in enumerate(questions, 1):
    print(f"{i}. {q}")

score = 0
print("到你了，请按次序作答：")
for i, ans in enumerate(answers, 1):
    user_answer = input(f"第{i}空：").strip()
    if user_answer == ans:
        print(f"第{i}空回答正确")
        score += 1
    else:
        print(f"第{i}空回答错误")
print(f"你的得分是：{score}/{len(answers)}")
```

```python
import random

# 输入题目数量并校验
while True:
    numbers = input("请输入题目数量：")
    if not numbers.isdigit():
        print("请输入100内的正整数：")
        continue
    numbers = int(numbers)
    if 0 < numbers <= 100:
        break
    print("必须输入100以内的正整数")

# 生成不出现负数结果的减法题
timus = []
for _ in range(numbers):
    a = random.randint(1, 100)
    b = random.randint(1, 100)
    if b > a:
        a, b = b, a
    timus.append((a, b))

# 正确答案
keys = [t[0] - t[1] for t in timus]

# 用户作答
answers = []
for t in timus:
    answers.append(input(f"{t[0]} - {t[1]} = "))

# 计算得分
score = 0
for idx in range(len(answers)):
    if not answers[idx].isdigit():
        continue
    if int(answers[idx]) == keys[idx]:
        score += 100 / len(keys)

print(f"得分：{score}")
```

### dict

```python
# 字典dict {k1:v1, k2:v2, k3:k3}
a_dict = {}     # 空字典，而非空set
b_dict = dict()     # 空字典
```

```python
# 字典的key，只能是不可修改的数据类型。value没有类型限制
a_dict = {
    1: "abc",
    0.99: 100,
    True: (1, 2, 3),   # True被当做了1来处理。  而False是0
    'abc': [100, 101, 102],
    (10, 11, 12): "abc"
}

print(a_dict)   # 字典的key只能是不可修改的数据类型。字典的value是没有限制的！
```

```python
# 字典常用的方法 keys(返回字典的key集合），values（字典的数据集合），items（返回字典的kv对）
a_dict = {
    "gejingyi": [91, 83, 87, 90],
    "xiajiayi": [81, 72, 68, 77],
    "xiayifei": [76, 85, 73, 92],
}
print(a_dict)
print('keys返回字典的所有key')
print(a_dict.keys())    # 结果为字典的keys对象。注意：不是list，但是一个序列组合

for k in a_dict.keys():
    print(k, a_dict[k])

print('values返回字典的数值value')
print(a_dict.values())
for v in a_dict.values():
    print(v)

print('items返回字典的kv对')
print(a_dict.items())
for k,v in a_dict.items():
    print(k, a_dict[k])
```

```python
# 根据字典的key访问value
print(a_dict['gejingyi'])   # 字典名[k]

# 如果key不存在，会导致报错！！！
# print(a_dict['不存在'])

# 为避免error可以通过get方法
print(a_dict.get('不存在'))    # 因为key不存在，返回None
# 可以指定当key不存在时的返回值
print(a_dict.get('不存在', 0))
print(a_dict.get('不存在', 100))
print(a_dict.get('不存在', '不存在哟～'))
```

```python
# 在字典中新增kv对。    最简单的办法是 字典名[key] = value
a_dict['X先生'] = [69,72,87,66]
print(a_dict)

# 如果key已经存咋了，则伤处方式直接修改了key对应的value
a_dict['xiayifei'] = [61, 58, 55, 61]   # 注意xiayifei已经存在，则直接修改了xiayifei对应的value

# 如果要避免已经存在的key被直接覆盖修改，可以使用setdefault方法
a_dict.setdefault('Y先生', [83, 79, 91, 93])      # 因为Y先生不存在，因此字典新增了Y先生
print(a_dict)

a_dict.setdefault('xieyifei', [83, 79, 91, 93])     # xieyifei已经存在，因此xieyifei对应的值不会被修改
print(a_dict)
```

```python
# 字典的fromkeys方法生产字典的key，value默认为None，但可以指定
a_dict = dict.fromkeys("abcd")
b_dict = dict.fromkeys(["xx", "yy", "zz"])
c_dict = dict.fromkeys(range(10))
print(a_dict) # value和None因为没有指定
print(b_dict)
print(c_dict)
```
