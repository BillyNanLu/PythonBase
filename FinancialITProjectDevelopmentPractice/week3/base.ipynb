### 练习1：模型

```python
# 有一个训练后的模型，准确率是60%
# 用9个该模型组成一个大模型，通过投票决定识别结果，那么，这个收的准确率是多少
# 用99个该模型组成一个大模型，准确率是多少呢

from math import comb

from PythonExercises.py010_sort_student_list_by_grade import students_sort


def ensemble_acc(n, p=0.6):
    k = n//2 + 1
    s = 0.0
    for i in range(k, n+1):
        s += comb(n, i) * (p**i) * ((1-p)**(n-i))
    return s

print("n=9  ->", ensemble_acc(9))    # ≈ 0.73343232
print("n=99 ->", ensemble_acc(99))   # ≈ 0.97806956
```

```python
import random

ACCUARY = 0.6
THE_NUMBERS = 9

TOTAL = 1000000
count = 0
for _ in range(TOTAL):
    results = [0] * THE_NUMBERS

    for idx in range(THE_NUMBERS):
        if random.random() < ACCUARY:
            results[idx] = 1

    if sum(results) >= int((THE_NUMBERS + 1) / 2):
        count += 1

print(f"使用{THE_NUMBERS}个模型得到的准确率是{count/TOTAL*100:.2f}%")
```

```python
# 集合类型 list tuple set dict
# 双色球： 1个蓝球（1，2，3，... ，16）
#         6个红球（1，2，3，... ，33）
# 模拟一个双色球售卖机，接受一个整数表示注数，然后随机给出彩票和费用

import random

zhushu = int(input("请输入您要购买的注数："))

price_per_ticket = 2  # 一注2元
total_price = zhushu * price_per_ticket

print(f"您购买 {zhushu} 注双色球，共需支付 {total_price} 元")
print("号码如下：")

tickets = []  # 存放每注彩票字典

for i in range(zhushu):
    # 6个不重复的红球（升序）
    red = sorted(random.sample(range(1, 34), 6))
    # 1个蓝球
    blue = random.randint(1, 16)

    ticket = {
        "注号": i + 1,
        "红球": red,
        "蓝球": blue,
        "价格": price_per_ticket
    }
    tickets.append(ticket)

# 打印每注字典
for t in tickets:
    print(t)
```

```python
PRICE = 2
# 接受一个正整数n，表示购买彩票的注数
while True:
    n = input("请输入一个正整数：")
    if not n.isdigit():
        print("必须要输入一个正整数")
        continue
    n = int(n)
    break

lotts = []

for _ in range(n):
    blue = random.choice(range(1, 17))
    red = random.sample(range(1, 34), k=6)
    lotts.append({"blue":blue, "red":red})

print("您购买的彩票是：")
for idx, lott in enumerate(lotts):
    print(f"第{idx+1}注：红球：", end="")
    for q in lott.get('red'):
        print(f"{q:4d}", end=" ")
    print(f"蓝球：{lott.get('blue')}")
print(f"请支付{n * PRICE}元")
```

### 综合练习：学生成绩管理系统

```python
# 综合练习 -- 学生成绩管理系统
# 录入学生的各科成绩。哟啊判断学生是否存在？学生的科目成绩是否已经存在？
# 对学生的科目成绩，可以进行增删改查

# 学生成绩管理系统（交互式）
# 数据结构：{学生姓名: {科目: 成绩}}

students = {}

def add_student(name):
    if name in students:
        print(f"学生 {name} 已存在！")
    else:
        students[name] = {}
        print(f"学生 {name} 添加成功。")

def add_score(name, subject, score):
    if name not in students:
        print(f"学生 {name} 不存在，请先添加学生。")
        return
    if subject in students[name]:
        print(f"{name} 的科目 {subject} 已经有成绩了，可用修改功能。")
        return
    students[name][subject] = score
    print(f"{name} 的 {subject} 成绩添加成功。")

def update_score(name, subject, score):
    if name not in students:
        print(f"学生 {name} 不存在。")
        return
    if subject not in students[name]:
        print(f"{name} 的 {subject} 没有成绩记录。")
        return
    students[name][subject] = score
    print(f"{name} 的 {subject} 成绩修改成功。")

def delete_score(name, subject):
    if name not in students:
        print(f"学生 {name} 不存在。")
        return
    if subject not in students[name]:
        print(f"{name} 的 {subject} 没有成绩记录。")
        return
    del students[name][subject]
    print(f"{name} 的 {subject} 成绩已删除。")

def show_scores(name):
    if name not in students:
        print(f"学生 {name} 不存在。")
        return
    if not students[name]:
        print(f"{name} 暂无成绩记录。")
        return
    print(f"{name} 的成绩：{students[name]}")

def show_all():
    if not students:
        print("暂无任何学生信息。")
        return
    for name, scores in students.items():
        print(f"{name}: {scores}")

# ========== 主循环 ==========
while True:
    print("\n=== 学生成绩管理系统 ===")
    print("1. 添加学生")
    print("2. 添加科目成绩")
    print("3. 修改科目成绩")
    print("4. 删除科目成绩")
    print("5. 查询某学生成绩")
    print("6. 查看全部学生")
    print("0. 退出")
    choice = input("请输入操作编号：")

    if choice == "1":
        name = input("请输入学生姓名：")
        add_student(name)

    elif choice == "2":
        name = input("请输入学生姓名：")
        subject = input("请输入科目：")
        score = float(input("请输入成绩："))
        add_score(name, subject, score)

    elif choice == "3":
        name = input("请输入学生姓名：")
        subject = input("请输入要修改的科目：")
        score = float(input("请输入新成绩："))
        update_score(name, subject, score)

    elif choice == "4":
        name = input("请输入学生姓名：")
        subject = input("请输入要删除的科目：")
        delete_score(name, subject)

    elif choice == "5":
        name = input("请输入学生姓名：")
        show_scores(name)

    elif choice == "6":
        show_all()

    elif choice == "0":
        print("系统已退出。")
        break
    else:
        print("无效的操作编号，请重新输入。")
```

```python
students = [] # 每个学生的属性：姓名  科目1成绩  科目2成绩...
subjects = ['C', 'Java', 'Python', 'English'] # 科目只需要记录名字

# 新增学生
print("新增学生：")
while True:
    name = input("请输入学生姓名：")
    if name in [s.get('name') for s in students]:
        print("该学生已经存在了。请重新输入学生姓名。")
        continue
    student = {'name':name}
    student.update(dict.fromkeys(subjects, "未定义"))
    students.append(student)

    print(f"已经录入了{len(students)}个学生")
    answer = input("是否还要继续新增学生？")
    if answer[0].upper() != 'Y':
        break

print(students)
# 录入学生成绩
while True:
    subject = input("请输入要录入的科目：")
    if subject not in subjects:
        print("没有这个科目，请重新输入")
        continue
    break

print(students)
```

```python
# 默认值传参 如果参数太多，函数的调用时的参数容易混乱 可以采用默认传参 减少参数传递
def hero(name, age, nation = "中国"): # 在函数定义时指定默认值
    print(f"{name}是一个{nation}的英雄，英雄故事已经流传了{age}年")

hero("孙悟空", 500) # nation采用了默认值，实参不需要传递
hero(age=10, name="美国队长", nation="美国") # 实参传递了nation，那么采用实参传递
```

```python
# 不定长参数 -- 参数的个数不能确定
# 形参钱有一个* 该参数类型是一个元组tuple 表示多个实参的组合
def get_sum(name, *numbers):
    total = 0
    for n in numbers:
        total += n
    print(f"{name}，你提交的数据之和为{total}")

get_sum("zhangsan", 65, 76, 88, 32, 96, 71)
```

```python
# 形参钱有2个* 表示该形参是一个字典 可以通过不定长的关键字类型传参
def show(name, **scores):
    print(f"{name}的各科成绩显示如下：")
    for k in scores.keys():
        print(k, scores.get(k))

show("张三", C=71, Java=79, Python=91)
```

```python
# 计算 1/2+1/3+2/5+3/8+5/13+8/21+... 前20项之和

# 第一个函数，计算当第一项是first，第二项是second
def fabo(first, second, n):
    if n == 1:
        return first
    if n == 2:
        return second
    count = 3
    a = first
    b = second
    while count <= n:
        a, b = b, a+b
        count += 1
    return b
```

## Python面向对象

```python
# python 面向对象编程的语言
# 类和对象
# python中一切都是对象（万物皆对象）任何对象都归属于某个类
# 类是生产对象的模版。类是对象的共性的抽象

class DemoClass():
    # 在init魔法函数中初始化的属性(self.)成为对象属性
    # self: 对象的本身   (java 中的 this)
    # 所有的对象方法。第一个参数都是self
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(f"name={self.name}, age={self.age}")
```

```python
d1 = DemoClass("zhangsan", 21)  # 调用了__init__() --对象方法的第一个参数self不用写，
d2 = DemoClass("zhangsan", 21)

d1.show()
d2.show()
```

```python
# 类属性
class Hero():
    nation = "中国"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(f"{self.name}是{Hero.nation}的英雄, 英雄故事已经传颂了{self.age}年")

    def __str__(self):
        return f"{Hero.nation}的英雄{self.name}"

    # repr对象
    def __repr__(self):
        return f"{Hero.nation}, {self.name}, {self.age}"

    # 可以用len
    def __len__(self):
        return self.age
```

```python
swk = Hero("孙悟空", 500)  # 利用类实例化对象

# 访问对象属性
print(swk.name)
print(swk.age)

# 访问对象方法
swk.show()

# 访问类属性nation
print(Hero.nation)  # 使用对象访问类属性，可以，但是不推荐!!!

print(swk)  # 如果期望使用print对象，那么，对象的类中，需要实现__str__模范函数

print(len(swk))
```

```python
class Students():
    def __init__(self, cname):
        self.classname = cname
        self.student_list = []

    def __str__(self):
        msg = f"{self.classname}班级有{len(self.student_list)}名学生："
        for s in self.student_list:
            msg += f"{s} "
        return msg

    def __repr__(self):
        return self.__str__()

    def __len__(self):
        return len(self.student_list)

    def append(self, name):
        self.student_list.append(name)

    def __getitem__(self, idx):
        if idx >= len(self.student_list):
            print("访问越界")
            return
        return self.student_list[idx]

    def __setitem__(self, idx, name):
        if idx >= len(self.student_list):
            print("没有那么多的学生！")
            return
        self.student_list[idx] = name
```

```python
class Person():
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        self.__name = name

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self, age):
        return self.__age

zhangsan = Person("zhangsan", 21)
print(zhangsan.name)
print(zhangsan.age)
zhangsan.name = "张三"
zhangsan.age = 31
print(zhangsan)
```

```python
# 定义两个武士：关于和张飞，可以给他们装备武器。让关于大战张飞
class Weapon():
    def __init__(self, name, ap):
        self.__name = name
        self.__ap = ap

    def __str__(self):
        return f"{self.__name}:{self.__ap}<UNK>"

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name):
        self.__name = name

    @property
    def ap(self):
        return self.__ap

    @ap.setter
    def ap(self, ap):
        self.__ap = ap

blade = Weapon("青龙偃月刀", 90)
spear = Weapon("丈八蛇矛",80)
print(blade.name)
print(spear.name)
```

```python
import time
import random

class Warriot():
    def __init__(self, name, hp=100, ap=50, dp=50, agile=20):
        self.name = name
        self.hp = hp
        self.agile = agile
        self.__ap = ap
        self.__dp = dp
        self.__weapon = None

    @property
    def ap(self):
        return self.__ap
    @ap.setter
    def ap(self, ap):
        self.__ap = ap

    @property
    def dp(self):
        return self.__dp
    @dp.setter
    def dp(self, dp):
        self.__dp = dp

    @property
    def alige(self):
        return self.__agile
    @alige.setter
    def agile(self, agile):
        self.__agile = agile

    @property
    def weapon(self):
        return self.__weapon
    @weapon.setter
    def weapon(self, weapon):
        self.__weapon = weapon

    def __str__(self):
        msg = f"{self.name}"
        if self.weapon:
            msg += f"手持{self.weapon.name}"
        msg += f': hp[{self.hp}] ap[{self.ap}] dp[{self.dp}] agile[{self.agile}]'
        return msg

    def wield(self, weapon):
        if self.weapon:
            print(f"{self.name}已经已经装备了{self.weapon.name}!")
            return
        print(f"{self.name}拿出{weapon.name}我在手中")
        self.weapon = weapon
        self.ap += weapon.ap

    def unwield(self, weapon):
        if not self.weapon:
            print(f"{self.name}本来就赤手空拳，还能放下什么")
            return
        print(f"{self.name}放下了手中持有的{self.weapon.name}!")
        self.weapon = None

    def fight(self, ob):
        print(f"{self.name}大喝一声：{ob.name}狗贼，拿命来！")
        while self.hp > 0 and ob.hp > 0:
            # 根据agile判定谁是攻击方，谁是防御方
            if random.randint(1, self.agile) > random.randint(1, ob.agile):
                attacker = self
                defender = ob
            else:
                attacker = ob
                defender = self

            if attacker.weapon:
                print(f"{attacker.name}挥动手中{weapon.name}向{defender.name}打去")
            else:
                print(f"{attacker.name}双拳交错，恶狠狠向{defender.name}打去")

            # 判定是否命中
            if random.randint(1, attacker.ap) > random.randint(1,defender.ap):
                # 攻击命中
                damage = random.randint(1, attacker.ap) - random.randint(1, defender.ap)
                defender.hp -= damage
                print(f"{defender.name}一击命中，顿时给{defender.name}造成了伤害")
            else:
                print(f"{defender.name}脚步轻盈，躲过了{attacker.name}的攻击")
            time.sleep(1)
        if self.hp > 0:
            print(f"{self.name}获得胜利")
        else:
            print(f"{ob.name}获得胜利")

guanyu = Warriot("关于",hp=120, ap=80, dp=70, agile=50)
zhangfei = Warriot("张飞",hp=125, ap=76, dp=76, agile=55)
print(guanyu)
print(zhangfei)
```
